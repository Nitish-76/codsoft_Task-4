# 3. Spliting the data into test and train

```python
X = slp1.drop(['gender','car purchase amount'], axis=1)
Y = slp1['car purchase amount']
print(X.shape,Y.shape)
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.2, random_state=3)
print(X.shape,X_train.shape, X_test.shape)
```

```python
regressor = RandomForestRegressor(n_estimators=100, random_state = 0)

regressor.fit(X_train, Y_train)

Giver2 = regressor.predict(X_test)

print('The mean Squared error:', mean_squared_error(Y_test, Giver2))
print('The mean Absolute error:', mean_absolute_error(Y_test, Giver2))
print('R2 Score of this model is:', r2_score(Y_test, Giver2))
print('Result by using RandomForestRegressor is this.')
```

```python
Model= LinearRegression()

Model.fit(X_train, Y_train)

Giver1 = Model.predict(X_test)


print('The mean Squared error:', mean_squared_error(Y_test, Giver1))
print('The mean Absolute error:', mean_absolute_error(Y_test, Giver1))
print('R2 Score of this model is:', r2_score(Y_test, Giver1))
print('Result by using LinearRegressor is this.')
```
